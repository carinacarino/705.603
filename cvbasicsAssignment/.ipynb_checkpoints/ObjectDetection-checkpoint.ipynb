```python
#!pip install opencv-python-headless
import cv2 as cv
import numpy as np
#!pip install matplotlib
import matplotlib.pyplot as plt 
#!pip install wandb
from wandb import Classes


def plot_cv_img(input_image):     
    """     
    Converts an image from BGR to RGB and plots     
    """   
    # change color channels order for matplotlib     
    plt.imshow(cv.cvtColor(input_image, cv.COLOR_BGR2RGB))          

    # For easier view, turn off axis around image     
    plt.axis('off')  
    # Must save prior to show - for show clears the image!
    #plt.savefig("DetectionOutput.jpg")
    plt.show()
```

## Set up layers and colors

```python
layer_name = net.getLayerNames()
output_layer = [layer_name[i - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes), 3))
```

## Fire up Model

```python
#note - 0.00392 = 1/250   416 is a standard square for yolo

blob = cv.dnn.blobFromImage(
    img,  0.00392, (416, 416), swapRB=True, crop=False)  # Fill in other fields to create image compatible with cv

#detect objects
net.setInput(blob)
outs = net.forward(output_layer)
```

## Set up image on screen  (all done for you)
* Note how to extract the identified objects and their assocated confidence.  You need that for part 2 of this assignment

```python
# Showing Information on the screen
class_ids = []
confidences = []
boxes = []
for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]
        if confidence > 0.5:
            # Object detection
            center_x = int(detection[0] * width)
            center_y = int(detection[1] * height)
            w = int(detection[2] * width)
            h = int(detection[3] * height)
            #cv.circle(img, (center_x, center_y), 10, (0, 255, 0), 2 )
            # Reactangle Cordinate
            x = int(center_x - w/2)
            y = int(center_y - h/2)
            boxes.append([x, y, w, h])
            confidences.append(float(confidence))
            class_ids.append(class_id)
           
print("Type: ", type(class_ids))

indexes = cv.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
print(indexes)

font = cv.FONT_HERSHEY_PLAIN
for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])
        print(f'Object: {label} with confidence of {confidences[i]:.2f}')
        color = colors[i]
        cv.rectangle(img, (x, y), (x + w, y + h), color, 2)
        cv.putText(img, label, (x, y + 30), font, 3, color, 3)

plot_cv_img(img)
```

## Effect of Image Size on Detection Confidence

```python
# Function that runs the object detection model and outputs the confidences for the object detected
def detect_objects(image, net, output_layers):
    blob = cv.dnn.blobFromImage(image, 0.00392, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)
    
    confidences = []
    
    # From the code above, our goal is to get the detection confidences that are over 0.5
    for out in outs:
        for detection in out:
            scores = detection[5:]
            confidence = scores[np.argmax(scores)]
            if confidence > 0.5:
                confidences.append(confidence)
    return confidences

# Load the image
original_image = cv.imread("pictures/bermuda.jpg")

# Setup layers and Colors
layer_name = net.getLayerNames()
output_layer = [layer_name[i - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes), 3))

# Set up scales that we can use to resize our original image
scales = np.linspace(0.001, 1.0, 100)

# Use median since mean can be skewed by outliers
median_confidences = []

for scale in scales:
    resized_image = cv.resize(original_image, None, fx=scale, fy=scale)
    confidences = detect_objects(resized_image, net, output_layer)
    median_confidence = np.median(confidences) if confidences else 0
    median_confidences.append(median_confidence)

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(scales, median_confidences, marker='o', linestyle='-', color='b')
plt.title('Median Detection Confidence vs. Image Size Scale Factor')
plt.xlabel('Image Size Scale Factor')
plt.ylabel('Median Confidence')
plt.grid(True)
plt.show()

```

### It looks like below the scale factor of 0.02, the model fails to detect any object with high confidence. Let's see what the resize image looks like at this scale factor:

```python
original_image = cv.imread("pictures/bermuda.jpg")
resized_img = cv.resize(original_image, None, fx=0.019,fy=0.019)
print(resized_img.shape)

plot_cv_img(resized_img) # around 1/10th of the training image height and 1/6th of the training image width
```

## Effect of Gaussian Noise on Detection Confidence

```python
#Function that adds gaussian noise to an image
def add_gaussian_noise(image, std):
    row, col, ch = image.shape
    mean = 0
    gauss = np.random.normal(mean, std, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    noisy = image + gauss
    noisy = np.clip(noisy, 0, 255).astype(np.uint8)
    return noisy

# Load original image
original_image = cv.imread('pictures/bermuda.jpg')
original_std = np.std(original_image)

std_scales = np.linspace(1, 6, 6)
median_confidences = []
 

for std_scale in std_scales:
    noisy_image = add_gaussian_noise(original_image, std_scale * original_std)
    confidences = detect_objects(noisy_image, net, output_layer)
    median_confidences.append(np.median(confidences) if confidences else 0)
    

# x-axis tick labels
std_labels = [f'{scale:.0f}σ' for scale in std_scales]

# Plotting
plt.figure(figsize=(10, 6))
plt.xticks(std_scales, labels=std_labels)
plt.plot(std_scales, median_confidences, marker='o', linestyle='-')
plt.title('Effect of Gaussian Noise on Detection Confidence')
plt.xlabel('Standard Deviation of Gaussian Noise')
plt.ylabel('Median Detection Confidence')
plt.grid(True)
plt.show()
```

The plot above shows that past 3σ, the model did not detect any object with high confidence. Which supports the rule of thumb to not choose standard devieation values that are greater than 3σ for gaussian filters.

## Effect of Salt and Pepper Noise on Detection Confidence

```python
# Add salt-and-pepper noise
def add_salt_and_pepper_noise(img, amount):
    noisy_img = random_noise(img, mode='s&p', amount=amount)
    
    noisy_img = np.array(255*noisy_img, dtype = 'uint8')
    
    return noisy_img

# Load original image
original_image = cv.imread('pictures/bermuda.jpg')

# Set noise probability amounts
amounts = np.linspace(0.01, 1.0, 100)
median_confidences = []

for amount in amounts:
    noisy_image = add_salt_and_pepper_noise(original_image, amount)
    confidences = detect_objects(noisy_image, net, output_layer)
    median_confidences.append(np.median(confidences) if confidences else 0)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(amounts, median_confidences, marker='o', linestyle='-')
plt.title('Effect of Salt-and-Pepper Noise on Detection Confidence')
plt.xlabel('Noise Amount')
plt.ylabel('Median Detection Confidence')
plt.grid(True)
plt.show()
```

From the plot above, we can see that around 0.5 noise probability, the model fails to detect any object with high confidence.
