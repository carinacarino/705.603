```python
import cv2
import os
import numpy as np

def extract_frames(video_path, output_dir, frame_rate):
    """
    Extracts and crops frames from a video at a specific frame rate using OpenCV,
    cropping the frames to a specific region before saving.

    Parameters:
    - video_path: Path to the video file.
    - output_dir: Directory where the extracted frames will be saved.
    - frame_rate: Frame rate at which to extract frames.
    """
    # Define the cropping region ((x1, y1), (x2, y2))
    crop_region = ((437, 386), (1592, 1077))
    
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error opening video file")
        return

    # Get the original video's frame rate and calculate frame extraction interval
    original_fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(original_fps / frame_rate)

    count = 0
    extracted_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Extract frames according to the specified frame rate
        if count % frame_interval == 0:
            # Crop the frame to the specified region
            cropped_frame = frame[crop_region[0][1]:crop_region[1][1], crop_region[0][0]:crop_region[1][0]]

            # Save the cropped frame
            frame_path = os.path.join(output_dir, f"frame_{extracted_count+1:04d}.tif")
            cv2.imwrite(frame_path, cropped_frame)
            extracted_count += 1

        count += 1

    cap.release()
    print(f"Extracted and cropped {extracted_count} frames to {output_dir}")
    
# Example usage
video_path = r"F:\workspace\705.603SP24\license_plate_recognition\moving_cars.mp4"
output_dir = r"F:\workspace\705.603SP24\license_plate_recognition\video_frames"
frame_rate = 1  # Specify the desired frame rate

extract_frames(video_path, output_dir, frame_rate)
```

```python
import cv2
import os
import numpy as np

def detect_objects(image, net, output_layers):
    """
    Detects objects using a given model and returns bounding boxes, confidences, and class IDs.
    """
    height, width = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    detections = net.forward(output_layers)
    
    boxes = []
    confidences = []
    class_ids = []
    for out in detections:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.9:
                center_x, center_y, w, h = (detection[0:4] * np.array([width, height, width, height])).astype('int')
                x, y = int(center_x - w / 2), int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
    return boxes, confidences, class_ids

def crop_and_save(image, box, output_dir, prefix, index):
    x, y, w, h = box
    # Ensure coordinates are within the image bounds
    x, y = max(x, 0), max(y, 0)
    w, h = min(w, image.shape[1] - x), min(h, image.shape[0] - y)

    cropped_image = image[y:y+h, x:x+w]
    if cropped_image.size == 0:
        print(f"Warning: Attempted to crop an empty region from the image. Skipping save for {prefix}_{index}.tif")
        return

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    file_path = os.path.join(output_dir, f"{prefix}_{index}.tif")
    cv2.imwrite(file_path, cropped_image)
    print(f"Image saved to {file_path}")


def process_images(input_dir, vehicle_net, lpr_net, vehicle_output_layers, lpr_output_layers, output_dir):
    image_files = [f for f in os.listdir(input_dir) if f.endswith(('.tif', '.tiff'))]
    
    for index, image_file in enumerate(image_files):
        image_path = os.path.join(input_dir, image_file)
        image = cv2.imread(image_path)
        
        # Detect vehicles
        vehicle_boxes, _, _ = detect_objects(image, vehicle_net, vehicle_output_layers)
        
        for v_index, vehicle_box in enumerate(vehicle_boxes):
            # Crop the vehicle image with bounds checking
            x_start, y_start = max(vehicle_box[0], 0), max(vehicle_box[1], 0)
            x_end, y_end = min(vehicle_box[0] + vehicle_box[2], image.shape[1]), min(vehicle_box[1] + vehicle_box[3], image.shape[0])
            vehicle_image = image[y_start:y_end, x_start:x_end]

            if vehicle_image.size == 0 or vehicle_image.shape[0] == 0 or vehicle_image.shape[1] == 0:
                print(f"Empty vehicle image at {image_path}, skipping license plate detection for vehicle {v_index}.")
                continue
            
            # Detect license plates within the vehicle bounding box
            plate_boxes, _, _ = detect_objects(vehicle_image, lpr_net, lpr_output_layers)
            
            if plate_boxes:
                # Only save the vehicle image if at least one license plate is detected
                crop_and_save(image, vehicle_box, output_dir, "vehicle", f"{index}_{v_index}")
                
                for p_index, plate_box in enumerate(plate_boxes):
                    # Crop and save the license plate image
                    crop_and_save(vehicle_image, plate_box, output_dir, "license_plate", f"{index}_{v_index}_{p_index}")



vehicle_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\object_detect_req\yolov3.weights", r"F:\workspace\705.603SP24\license_plate_recognition\object_detect_req\yolov3.cfg")
lpr_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.weights", r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.cfg")


frames_dir = r"F:\workspace\705.603SP24\license_plate_recognition\video_frames"
output_dir = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_vehicles"

vehicle_layer_names = vehicle_net.getLayerNames()
vehicle_output_layers = [vehicle_layer_names[i - 1] for i in vehicle_net.getUnconnectedOutLayers()]

lpr_layer_names = lpr_net.getLayerNames()
lpr_output_layers = [lpr_layer_names[i - 1] for i in lpr_net.getUnconnectedOutLayers()]

frames_dir = r"F:\workspace\705.603SP24\license_plate_recognition\video_frames"
output_dir = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_vehicles"
            
process_images(frames_dir, vehicle_net, lpr_net, vehicle_output_layers, lpr_output_layers, output_dir)
```

```python
import cv2
import csv
from pathlib import Path
import os
import numpy as np

# Load the pre-trained LPR model
lpr_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.weights", 
                          r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.cfg")

def track_license_plates(frames_dir, conf_threshold=0.5, csv_file="license_plates.csv"):
    """
    Tracks license plate objects in a directory of TIF video frames using a pre-trained LPR model.
    Returns a list of frames and a dictionary containing the last seen frame and bounding box
    for each license plate object. Creates a new CSV file if it doesn't exist.
    """
    frames = []
    tracked_objects = {}

    # Load the TIF frames from the directory
    for frame_path in sorted(Path(frames_dir).glob("*.tif")):
        frame = cv2.imread(str(frame_path), cv2.IMREAD_UNCHANGED)
        if frame is None:
            print(f"Warning: Failed to load frame {frame_path}")
        frames.append(frame)

    for frame_num, frame in enumerate(frames):
        height, width = frame.shape[:2]  # Get the height and width of the frame

        # Detect license plates in the current frame
        blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)
        lpr_net.setInput(blob)
        layer_outputs = lpr_net.forward(lpr_net.getUnconnectedOutLayersNames())

        # Loop over the detected license plates
        detections_in_frame = 0
        for output in layer_outputs:
            for detection in output:
                scores = detection[5:]
                confidence = scores.max()

                # Check if the object has high confidence
                if confidence > conf_threshold:
                    detections_in_frame += 1

                    # Calculate the bounding box coordinates
                    center_x, center_y, w, h = (detection[0:4] * np.array([width, height, width, height])).astype('int')
                    x, y = int(center_x - w / 2), int(center_y - h / 2)

                    # Construct a unique ID for the license plate
                    object_id = f"{x}_{y}_{w}_{h}"

                    # Update the tracked objects dictionary with the last seen frame and bounding box
                    tracked_objects[object_id] = (frame_num, (x, y, w, h))

        print(f"Frame {frame_num}: {detections_in_frame} license plates detected")

    # Create a new CSV file if it doesn't exist
    try:
        with open(csv_file, 'x', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(["license_plate_object_id", "last_seen_frame", "bounding_box"])
    except FileExistsError:
        pass

    # Write the tracked objects to the CSV file
    with open(csv_file, 'a', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for object_id, (last_seen_frame, bbox) in tracked_objects.items():
            writer.writerow([object_id, last_seen_frame, bbox])

    return frames, tracked_objects

def crop_and_save_license_plates(frames, tracked_objects, output_dir):
    """
    Crops and saves the license plate regions from the frames based on the bounding boxes in the tracked_objects dictionary.
    """
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    for object_id, (last_seen_frame, bbox) in tracked_objects.items():
        frame = frames[last_seen_frame]
        x, y, w, h = bbox

        # Ensure coordinates are within the image bounds
        x, y = max(x, 0), max(y, 0)
        w, h = min(w, frame.shape[1] - x), min(h, frame.shape[0] - y)

        # Crop the license plate region
        cropped_plate = frame[y:y+h, x:x+w]

        # Check if the cropped region is empty
        if cropped_plate.size == 0:
            print(f"Warning: Attempted to crop an empty region from the image. Skipping save for {object_id}_{last_seen_frame}.tif")
            continue

        # Construct the output file path
        cropped_plate_path = os.path.join(output_dir, f"{object_id}_{last_seen_frame}.tif")

        # Save the cropped license plate image
        cv2.imwrite(cropped_plate_path, cropped_plate)
        print(f"Image saved to {cropped_plate_path}")

# Call the track_license_plates function
frames, tracked_objects = track_license_plates(r"F:\workspace\705.603SP24\license_plate_recognition\video_frames", conf_threshold=0.5)

# Set the working directory to the location where the CSV file is located
os.chdir(r"F:\workspace\705.603SP24")

# Specify the output directory for the cropped license plate images
output_dir = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_licenseplates"

# Call the crop_and_save_license_plates function
crop_and_save_license_plates(frames, tracked_objects, output_dir)
```

```python
import cv2

# Initialize video capture
cap = cv2.VideoCapture("F:\workspace\705.603SP24\license_plate_recognition\moving_cars.mp4")

lpr_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.weights", 
                          r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.cfg")

# Dictionary to keep track of the last seen frame and bounding box for each license plate
last_seen_info = {}

def perform_detection(image, net, conf_threshold=0.5):
    (H, W) = image.shape[:2]
    # Create a blob from the input frame and then perform a forward pass of the YOLO object detector,
    # giving us our bounding boxes and associated probabilities.
    blob = cv2.dnn.blobFromImage(image, 1 / 255.0, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    layer_outputs = net.forward(net.getUnconnectedOutLayersNames())

    boxes = []
    confidences = []
    # Loop over each of the layer outputs
    for output in layer_outputs:
        # Loop over each of the detections
        for detection in output:
            # Extract the class ID and confidence (i.e., probability) of the current object detection
            scores = detection[5:]
            classID = np.argmax(scores)
            confidence = scores[classID]

            # Filter out weak predictions by ensuring the detected probability is greater than the minimum probability
            if confidence > conf_threshold:
                # Scale the bounding box coordinates back relative to the size of the image,
                # YOLO returns the center (x, y)-coordinates of the bounding box followed by the boxes' width and height
                box = detection[0:4] * np.array([W, H, W, H])
                (centerX, centerY, width, height) = box.astype("int")

                # Use the center (x, y)-coordinates to derive the top and left corner of the bounding box
                x = int(centerX - (width / 2))
                y = int(centerY - (height / 2))

                # Update our list of bounding box coordinates, confidences, and class IDs
                boxes.append([x, y, int(width), int(height)])
                confidences.append(float(confidence))

    # Apply non-maxima suppression to suppress weak, overlapping bounding boxes
    idxs = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, conf_threshold)

    # Ensure at least one detection exists
    final_boxes = []
    if len(idxs) > 0:
        # Loop over the indexes we are keeping
        for i in idxs.flatten():
            # Extract the bounding box coordinates
            final_boxes.append(boxes[i])

    return final_boxes



import cv2

def main(video_path):
    cap = cv2.VideoCapture(video_path)
    trackers = []
    last_seen_info = {}  # Key: tracker index, Value: (last_seen_frame, bbox)

    frame_num = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Assuming perform_detection is your real detection function
        new_detections = perform_detection(frame, lpr_net)

        if frame_num == 0 or not trackers:  # Initial detection
            for bbox in new_detections:
                tracker = cv2.TrackerCSRT_create()  # Using CSRT tracker
                trackers.append(tracker)
                tracker.init(frame, tuple(bbox))
        else:
            # Update existing trackers
            for i, tracker in enumerate(trackers):
                success, bbox = tracker.update(frame)
                if success:
                    # Update last seen info
                    last_seen_info[i] = (frame_num, bbox)
        
        frame_num += 1

    # Cleanup
    cap.release()
    cv2.destroyAllWindows()

    # Output the last seen frame and bbox for each tracker
    for tracker_id, (last_seen_frame, bbox) in last_seen_info.items():
        print(f"Tracker {tracker_id}: Last Seen Frame: {last_seen_frame}, BBox: {bbox}")

if __name__ == "__main__":
    video_path = r"F:\workspace\705.603SP24\license_plate_recognition\moving_cars.mp4"  # Update this to your video path
    main(video_path)

```

```python
import matplotlib.pyplot as plt

def display_original_and_cropped(original_image_path, coco_cropped_path, yolo_cropped_path):
    # Load the original image
    original_image = cv2.imread(original_image_path)
    
    # Load the cropped images
    coco_cropped_image = cv2.imread(coco_cropped_path)
    yolo_cropped_image = cv2.imread(yolo_cropped_path)
    
    # Convert images from BGR to RGB for matplotlib display
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    coco_cropped_image_rgb = cv2.cvtColor(coco_cropped_image, cv2.COLOR_BGR2RGB)
    yolo_cropped_image_rgb = cv2.cvtColor(yolo_cropped_image, cv2.COLOR_BGR2RGB)
    
    # Create a figure and subplots
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Display the original image
    axes[0].imshow(original_image_rgb)
    axes[0].set_title("Original Image")
    axes[0].axis("off")
    
    # Display the COCO cropped image
    axes[1].imshow(coco_cropped_image_rgb)
    axes[1].set_title("COCO Cropped Image")
    axes[1].axis("off")
    
    # Display the YOLO cropped image
    axes[2].imshow(yolo_cropped_image_rgb)
    axes[2].set_title("YOLO Cropped Image")
    axes[2].axis("off")

    plt.tight_layout()
    plt.show()
    
    
```

```python
import cv2
import pytesseract

def extract_license_plate_number(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray_image = cv2.resize(gray_image, (333, 105))
    
    # Preprocess the image
    adaptive_thresh_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 65, 5)
    inverted_image = cv2.bitwise_not(adaptive_thresh_image)
    
    custom_config = r'-l eng --psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    
    license_plate_number = pytesseract.image_to_string(inverted_image, config=custom_config)
    
    license_plate_number = license_plate_number.strip()
    
    plt.imshow(inverted_image, cmap='gray')
    plt.title('Processed Image')
    plt.axis('off')
    plt.show()
    
    return license_plate_number

# Example
cropped_image_path = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_vehicles\license_plate_15_8_0.tif"

cropped_image = cv2.imread(cropped_image_path)
license_plate_number = extract_license_plate_number(cropped_image)

print("Extracted License Plate Number:", license_plate_number)
```

```python
import os
import pytesseract
import cv2

# Set the TESSDATA_PREFIX environment variable
os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'

def detect_license_plate(image_path):
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray_image = cv2.resize(gray_image, (333, 105))

    adaptive_thresh_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 65, 5)
    inverted_image = cv2.bitwise_not(adaptive_thresh_image)

    data = pytesseract.image_to_string(inverted_image, config='--psm 8')
    data = data.strip()

    if data:
        custom_config = r'--psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
        confidence_data = pytesseract.image_to_data(inverted_image, config=custom_config, output_type=pytesseract.Output.DICT)
        conf_values = [float(val) for val in confidence_data['conf'] if val != '-1']
        if conf_values:
            string_confidence = min(conf_values)
            return data, string_confidence
        else:
            return data, 0.0
    else:
        return None, None

def process_license_plate_directory(directory_path):
    for filename in os.listdir(directory_path):
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".tif"):
            image_path = os.path.join(directory_path, filename)
            license_plate, confidence = detect_license_plate(image_path)

            if license_plate:
                print(f"Filename: {filename}")
                print("Predicted License Plate:", license_plate)
                print("String Confidence Score: {:.2f}%".format(confidence))
                print("---")
            else:
                print(f"No license plate detected in {filename}")
                print("---")

# Example usage
directory_path = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_licenseplates"
process_license_plate_directory(directory_path)
```
