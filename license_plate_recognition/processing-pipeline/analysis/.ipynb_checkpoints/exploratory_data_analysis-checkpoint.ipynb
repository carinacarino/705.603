```python
def extract_frames(video_path, output_dir, frame_rate):
    """
    Extracts and crops frames from a video at a specific frame rate using OpenCV,
    cropping the frames to a specific region, enhancing their contrast, and then
    enlarging them to double their size before saving.

    Parameters:
    - video_path: Path to the video file.
    - output_dir: Directory where the extracted frames will be saved.
    - frame_rate: Frame rate at which to extract frames.
    """
    # Define the cropping region ((x1, y1), (x2, y2))
    crop_region = ((437, 635), (1592, 1077))
    
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error opening video file")
        return

    # Get the original video's frame rate and calculate frame extraction interval
    original_fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(original_fps / frame_rate)

    count = 0
    extracted_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Extract frames according to the specified frame rate
        if count % frame_interval == 0:
            # Crop the frame to the specified region
            cropped_frame = frame[crop_region[0][1]:crop_region[1][1], crop_region[0][0]:crop_region[1][0]]
            
            # Increase the contrast of the cropped frame
            alpha = 1.5 # Contrast control (>1.0 increases contrast, <1.0 decreases contrast)
            beta = 0    # Brightness control (not used here)
            contrast_enhanced_frame = cv2.convertScaleAbs(cropped_frame, alpha=alpha, beta=beta)
            
            # Enlarge the contrast-enhanced frame to double its size
            enlarged_frame = cv2.resize(contrast_enhanced_frame, None, fx=3, fy=3, interpolation=cv2.INTER_LINEAR)

            # Save the enlarged and contrast-enhanced frame
            frame_path = os.path.join(output_dir, f"frame_{extracted_count+1:04d}.tif")
            cv2.imwrite(frame_path, enlarged_frame)
            extracted_count += 1

        count += 1

    cap.release()
    print(f"Extracted, cropped, contrast-enhanced, and enlarged {extracted_count} frames to {output_dir}")

# Example usage
video_path = "F:\\workspace\\705.603SP24\\license_plate_recognition\\moving_cars.mp4"
output_dir = "F:\\workspace\\705.603SP24\\license_plate_recognition\\video_frames"
frame_rate = 2  # Specify the desired frame rate

#extract_frames(video_path, output_dir, frame_rate)

```

```python

from cv2 import dnn_superres

def extract_frames(video_path, output_dir, frame_rate):
    """
    Extracts and crops frames from a video at a specific frame rate using OpenCV,
    cropping the frames to a specific region, enhancing their contrast, applying
    super-resolution, and then saving the frames.

    Parameters:
    - video_path: Path to the video file.
    - output_dir: Directory where the extracted frames will be saved.
    - frame_rate: Frame rate at which to extract frames.
    """
    # Define the cropping region ((x1, y1), (x2, y2))
    crop_region = ((437, 635), (1592, 1077))
    
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error opening video file")
        return

    # Get the original video's frame rate and calculate frame extraction interval
    original_fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(original_fps / frame_rate)

    # Create an SR object
    sr = dnn_superres.DnnSuperResImpl_create()

    # Read the desired model
    path = r"F:\workspace\705.603SP24\license_plate_recognition\EDSR_x3.pb"
    sr.readModel(path)

    # Set the desired model and scale to get correct pre- and post-processing
    sr.setModel("edsr", 3)

    count = 0
    extracted_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Extract frames according to the specified frame rate
        if count % frame_interval == 0:
            # Crop the frame to the specified region
            cropped_frame = frame[crop_region[0][1]:crop_region[1][1], crop_region[0][0]:crop_region[1][0]]
            
            # Increase the contrast of the cropped frame
            alpha = 1.5 # Contrast control (>1.0 increases contrast, <1.0 decreases contrast)
            beta = 0    # Brightness control (not used here)
            contrast_enhanced_frame = cv2.convertScaleAbs(cropped_frame, alpha=alpha, beta=beta)
            
            # Apply super-resolution to the contrast-enhanced frame
            sr_frame = sr.upsample(contrast_enhanced_frame)

            # Save the super-resolution frame
            frame_path = os.path.join(output_dir, f"frame_{extracted_count+1:04d}.tif")
            cv2.imwrite(frame_path, sr_frame)
            extracted_count += 1

        count += 1

    cap.release()
    print(f"Extracted, cropped, contrast-enhanced, and super-resolved {extracted_count} frames to {output_dir}")

# Example usage
video_path = "F:\\workspace\\705.603SP24\\license_plate_recognition\\moving_cars.mp4"
output_dir = "F:\\workspace\\705.603SP24\\license_plate_recognition\\video_frames2"
frame_rate = 2  # Specify the desired frame rate

#extract_frames(video_path, output_dir, frame_rate)
```

```python
import cv2
import os


def detect_objects(image, net, output_layers):
    """
    Detects objects using a given model and returns bounding boxes, confidences, and class IDs.
    """
    height, width = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    detections = net.forward(output_layers)
    
    boxes = []
    confidences = []
    class_ids = []
    for out in detections:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.8:
                center_x, center_y, w, h = (detection[0:4] * np.array([width, height, width, height])).astype('int')
                x, y = int(center_x - w / 2), int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
    return boxes, confidences, class_ids

def crop_and_save(image, box, output_dir, prefix, index):
    x, y, w, h = box
    # Ensure coordinates are within the image bounds
    x, y = max(x, 0), max(y, 0)
    w, h = min(w, image.shape[1] - x), min(h, image.shape[0] - y)

    cropped_image = image[y:y+h, x:x+w]
    if cropped_image.size == 0:
        print(f"Warning: Attempted to crop an empty region from the image. Skipping save for {prefix}_{index}.tif")
        return

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    file_path = os.path.join(output_dir, f"{prefix}_{index}.tif")
    cv2.imwrite(file_path, cropped_image)
    print(f"Image saved to {file_path}")


def process_images(input_dir, vehicle_net, lpr_net, vehicle_output_layers, lpr_output_layers, output_dir):
    image_files = [f for f in os.listdir(input_dir) if f.endswith(('.tif', '.tiff'))]
    
    for index, image_file in enumerate(image_files):
        image_path = os.path.join(input_dir, image_file)
        image = cv2.imread(image_path)
        
        # Detect vehicles
        vehicle_boxes, _, _ = detect_objects(image, vehicle_net, vehicle_output_layers)
        
        for v_index, vehicle_box in enumerate(vehicle_boxes):
            # Crop the vehicle image with bounds checking
            x_start, y_start = max(vehicle_box[0], 0), max(vehicle_box[1], 0)
            x_end, y_end = min(vehicle_box[0] + vehicle_box[2], image.shape[1]), min(vehicle_box[1] + vehicle_box[3], image.shape[0])
            vehicle_image = image[y_start:y_end, x_start:x_end]

            if vehicle_image.size == 0 or vehicle_image.shape[0] == 0 or vehicle_image.shape[1] == 0:
                print(f"Empty vehicle image at {image_path}, skipping license plate detection for vehicle {v_index}.")
                continue
            
            # Detect license plates within the vehicle bounding box
            plate_boxes, _, _ = detect_objects(vehicle_image, lpr_net, lpr_output_layers)
            
            if plate_boxes:
                # Only save the vehicle image if at least one license plate is detected
                crop_and_save(image, vehicle_box, output_dir, "vehicle", f"{index}_{v_index}")
                
                for p_index, plate_box in enumerate(plate_boxes):
                    # Crop and save the license plate image
                    crop_and_save(vehicle_image, plate_box, output_dir, "license_plate", f"{index}_{v_index}_{p_index}")
                    
                    



vehicle_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\object_detect_req\yolov3.weights", r"F:\workspace\705.603SP24\license_plate_recognition\object_detect_req\yolov3.cfg")
lpr_net = cv2.dnn.readNet(r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.weights", r"F:\workspace\705.603SP24\license_plate_recognition\yolo_lpr\lpr-yolov3.cfg")


vehicle_layer_names = vehicle_net.getLayerNames()
vehicle_output_layers = [vehicle_layer_names[i - 1] for i in vehicle_net.getUnconnectedOutLayers()]

lpr_layer_names = lpr_net.getLayerNames()
lpr_output_layers = [lpr_layer_names[i - 1] for i in lpr_net.getUnconnectedOutLayers()]

frames_dir = r"F:\workspace\705.603SP24\license_plate_recognition\video_frames2"
output_dir = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_vehicles"
            
process_images(frames_dir, vehicle_net, lpr_net, vehicle_output_layers, lpr_output_layers, output_dir)
```

```python
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt

def deblur_image(image):
    # Ensure image is grayscale before applying Gaussian blur for deblurring
    if len(image.shape) > 2:  # Image is not grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:  # Image is already grayscale
        gray_image = image
    gaussian_blur = cv2.GaussianBlur(gray_image, (0, 0), 3)
    sharpened_image = cv2.addWeighted(gray_image, 1.5, gaussian_blur, -0.5, 0)
    return sharpened_image

def remove_blue_logo(image):
    # Make sure this function is called with a BGR image
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_blue = np.array([100, 150, 50])
    upper_blue = np.array([140, 255, 255])
    mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
    image[mask > 0] = (255, 255, 255)  # Replace blue with white
    return image

def extract_license_plate_number(image):
    # First, remove the blue logo (make sure `image` is in BGR format here)
    logo_removed_image = remove_blue_logo(image)
    
    # Now, convert to grayscale for further processing
    gray_image = cv2.cvtColor(logo_removed_image, cv2.COLOR_BGR2GRAY)
    
    # Deblur the grayscale image
    deblurred_image = deblur_image(gray_image)
    
    # Resize the deblurred image
    resized_image = cv2.resize(deblurred_image, (300, 100), interpolation=cv2.INTER_LINEAR)
    
    # Preprocess the image
    adaptive_thresh_image = cv2.adaptiveThreshold(resized_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,35, 1)
    # Inverting the adaptive threshold image to create a binary mask
    inverted_image = cv2.bitwise_not(adaptive_thresh_image)
    
    # Find all connected components (the background label is 0)
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(inverted_image, connectivity=8)
    
    # Background is the largest component so we start comparing from the second component
    largest_component_label = 1 + np.argmax(stats[1:, cv2.CC_STAT_AREA])
    
    # Create a mask for the largest connected component
    largest_component_mask = (labels == largest_component_label).astype(np.uint8) * 255
    
    # Remove the largest connected component by setting it to zero (black)
    inverted_image[largest_component_mask > 0] = 0
    
    # Continue with your existing code...
    custom_config = r'--psm 8 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    license_plate_number = pytesseract.image_to_string(inverted_image, config=custom_config)
    
    
    license_plate_number = license_plate_number.strip()
    
    plt.imshow(inverted_image, cmap='gray')
    plt.title('Processed Image with Blue Logo Removed')
    plt.axis('off')
    plt.show()
    
    return license_plate_number



# cropped_image_path = r"F:\workspace\705.603SP24\license_plate_recognition\cropped_licenseplates\license_plate_53_0_0.tif"
# # Assuming you have 'cropped_image_path' defined and the image is loaded correctly
# cropped_image = cv2.imread(cropped_image_path)
# license_plate_number = extract_license_plate_number(cropped_image)
# 
# print("Extracted License Plate Number:", license_plate_number)

```

```python
import cv2
import os
import pytesseract
import matplotlib as plt
from PIL import Image

def binarize_lp(image):
    img_lp = cv2.resize(image, (666, 150))
    img_gray_lp = cv2.cvtColor(img_lp, cv2.COLOR_BGR2GRAY)
    _, img_binary_lp = cv2.threshold(img_gray_lp, 150, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    img_binary_lp = cv2.erode(img_binary_lp, (3,3))
    img_binary_lp = cv2.dilate(img_binary_lp, (3,3))
    
    LP_WIDTH = img_binary_lp.shape[0]
    LP_HEIGHT = img_binary_lp.shape[1]
    
    img_binary_lp[0:3,:] = 255
    img_binary_lp[:,0:3] = 255
    img_binary_lp[72:75,:] = 255
    img_binary_lp[:,330:333] = 255
    
    return img_binary_lp

def extract_license_plate_text(cropped_image_path):
    img = cv2.imread(cropped_image_path)
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    
    contours, _ = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    largest_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(largest_contour)
    
    cropped_image_simple = img[y:y+h, x:x+w]
    
    bw = binarize_lp(cropped_image_simple)
    
    os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'
    
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
    custom_config = r'--psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    
    text = pytesseract.image_to_string(bw, config=custom_config)
    
    return text
```
