**(PART 1) Converting bounding box formats**

As part of your Data Pipeline class, using the sample license plate recognition dataset, write two functions that takes in a single file (.xml) and converts:

- PASCAL VOC format to YOLO format (e.g., `convert_pascal_to_yolo_format(filename='<path/to/file>/testfile.xml')`)
- PASCAL VOC format to COCO format `(e.g., convert_pascal_to_coco_format(filename='<path/to/file>/testfile.xml')`)

Then, using these functions above, write a script that sequentially parses all PASCAL VOC annotation format .xml files in a directory and produce:
- YOLO formatted annotations
- COCO formatted annotations
___

```python
import os
import xml.etree.ElementTree as ET
import json

class DataPipeline:
    def __init__(self, xml_directory: str, yolo_output_directory: str, coco_output_file: str):
        self.xml_directory = xml_directory
        self.yolo_output_directory = yolo_output_directory
        self.coco_output_file = coco_output_file
        self.coco_dataset = {
            "images": [],
            "annotations": [],
            "categories": [{"id": 1, "name": "object"}]  
        }
        self.annotation_id = 1 

    def convert_xml_to_yolo(self, xml_file_path: str):
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        width = int(root.find('size/width').text)
        height = int(root.find('size/height').text)
        file_name = root.find('filename').text
        base_name = os.path.splitext(file_name)[0]

        yolo_data = []
        for obj in root.findall('object'):
            bndbox = obj.find('bndbox')
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)

            # Normalize the bounding box values
            x_center = ((xmin + xmax) / 2) / width
            y_center = ((ymin + ymax) / 2) / height
            w = (xmax - xmin) / width
            h = (ymax - ymin) / height

 
            yolo_data.append(f"0 {x_center} {y_center} {w} {h}")

        # Ensure output directory exists
        if not os.path.exists(self.yolo_output_directory):
            os.makedirs(self.yolo_output_directory)

        # Save the YOLO formatted annotations
        with open(os.path.join(self.yolo_output_directory, f"{base_name}.txt"), 'w') as file:
            for annotation in yolo_data:
                file.write(annotation + "\n")


    def convert_xml_to_coco(self, xml_file_path: str):
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        width = int(root.find('size/width').text)
        height = int(root.find('size/height').text)
        file_name = root.find('filename').text

        # Add image information to COCO dataset
        image_id = len(self.coco_dataset["images"]) + 1
        self.coco_dataset["images"].append({
            "id": image_id,
            "width": width,
            "height": height,
            "file_name": file_name
        })

        # Parse each object in the XML file
        for obj in root.findall('object'):
            bndbox = obj.find('bndbox')
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)

            # Add annotation information to COCO dataset
            self.coco_dataset["annotations"].append({
                "id": self.annotation_id,
                "image_id": image_id,
                "category_id": 1,  # Single generic category
                "bbox": [xmin, ymin, xmax - xmin, ymax - ymin],
                "area": (xmax - xmin) * (ymax - ymin),
                "iscrowd": 0
            })
            self.annotation_id += 1

    def process_all_xml_files(self):
        for xml_file in os.listdir(self.xml_directory):
            if xml_file.endswith('.xml'):
                xml_file_path = os.path.join(self.xml_directory, xml_file)
                self.convert_xml_to_yolo(xml_file_path)
                self.convert_xml_to_coco(xml_file_path)

        # After processing all files, save the COCO annotations to a JSON file
        with open(self.coco_output_file, 'w') as f:
            json.dump(self.coco_dataset, f, indent=4)


xml_directory = r'F:\workspace\705.603SP24\license_plate_recognition\license_plate_dataset'
yolo_output_directory = r'F:\workspace\705.603SP24\license_plate_recognition\annotations\YOLO_annotations'
coco_output_file = r'F:\workspace\705.603SP24\license_plate_recognition\annotations\coco_annotations.json'
pipeline = DataPipeline(xml_directory, yolo_output_directory, coco_output_file)
pipeline.process_all_xml_files()
```

```python
import json
import cv2
import numpy as np
import os

def crop_image_from_bbox(image, bbox):

    x_min, y_min, width, height = bbox
    x_max = x_min + width
    y_max = y_min + height
    cropped_image = image[y_min:y_max, x_min:x_max]
    return cropped_image

def crop_image_from_bbox_yolo(image, bbox):

    image_height, image_width = image.shape[:2]
    x_center, y_center, width, height = bbox

    # Convert normalized coordinates to pixel coordinates
    x_center *= image_width
    y_center *= image_height
    width *= image_width
    height *= image_height

    # Calculate the top-left and bottom-right coordinates of the bounding box
    x_min = int(x_center - width / 2)
    y_min = int(y_center - height / 2)
    x_max = int(x_center + width / 2)
    y_max = int(y_center + height / 2)

    # Crop the image using the pixel coordinates
    cropped_image = image[y_min:y_max, x_min:x_max]
    return cropped_image

def find_annotation_for_image(image_filename, annotations_data):

    for img in annotations_data['images']:
        if img['file_name'] == image_filename:
            image_id = img['id']
            for ann in annotations_data['annotations']:
                if ann['image_id'] == image_id:
                    return ann
    return None
  
def process_images(annotation_path, annotation_format, image_directory, cropped_directory):
    if annotation_format == 'COCO':
        # Load COCO annotations
        with open(annotation_path) as f:
            coco_data = json.load(f)

        # Process COCO annotations
        for img in coco_data['images']:
            image_filename = img['file_name']
            image_path = os.path.join(image_directory, image_filename)
            original_image = cv2.imread(image_path)

            if original_image is None:
                print(f"Failed to load image: {image_path}")
                continue

            annotation = find_annotation_for_image(image_filename, coco_data)
            if annotation:
                bbox = annotation['bbox']
                cropped_image = crop_image_from_bbox(original_image, bbox)

                cropped_filename = f"cropped_{annotation_format}_{image_filename}"
                cv2.imwrite(os.path.join(cropped_directory, cropped_filename), cropped_image)

    elif annotation_format == 'YOLO':
        # Get the list of annotation filenames in the directory
        annotation_filenames = [filename for filename in os.listdir(annotation_path) if filename.lower().endswith('.txt')]

        # Process YOLO annotations
        for annotation_filename in annotation_filenames:
            annotation_file_path = os.path.join(annotation_path, annotation_filename)
            image_filename = os.path.splitext(annotation_filename)[0] + '.jpeg'
            image_path = os.path.join(image_directory, image_filename)
            original_image = cv2.imread(image_path)

            if original_image is None:
                print(f"Failed to load image: {image_path}")
                continue

            with open(annotation_file_path, 'r') as f:
                annotation_data = f.read().strip().split(' ')

            bbox = list(map(float, annotation_data[1:]))
            cropped_image = crop_image_from_bbox_yolo(original_image, bbox)

            cropped_filename = f"cropped_{annotation_format}_{image_filename}"
            cv2.imwrite(os.path.join(cropped_directory, cropped_filename), cropped_image)

    # Create the cropped directory if it doesn't exist
    os.makedirs(cropped_directory, exist_ok=True)
```

```python
import cv2
import pytesseract

def extract_license_plate_number(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Preprocess the image
    adaptive_thresh_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 3)
    inverted_image = cv2.bitwise_not(adaptive_thresh_image)
    
    custom_config = r'-l eng --psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    
    license_plate_number = pytesseract.image_to_string(inverted_image, config=custom_config)
    
    license_plate_number = license_plate_number.strip()
    
    plt.imshow(inverted_image, cmap='gray')
    plt.title('Processed Image')
    plt.axis('off')
    plt.show()
    
    return license_plate_number

cropped_image_path = r"F:/workspace/705.603SP24/license_plate_recognition/cropped_images/YOLO/cropped_YOLO_N1.jpeg"

cropped_image = cv2.imread(cropped_image_path)
license_plate_number = extract_license_plate_number(cropped_image)

print("Extracted License Plate Number:", license_plate_number)
```
